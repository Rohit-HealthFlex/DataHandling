```python
import pandas as pd
import numpy as np
```

```python
def read_data(filename, req_cols=None):
    df = pd.read_csv(filename)
    df = df.reset_index()
    df.iloc[:, :] = df.iloc[:, :].shift(axis=1)
    df = df.drop(["index"], axis=1)
    return df
```
